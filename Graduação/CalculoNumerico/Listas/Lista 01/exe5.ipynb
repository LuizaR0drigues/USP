```python
class No:
     def __init__(self, x,y, dado):
        self.coord = [x,y]
        self.conex = []
        self.info = dado

     def conecta_no(self, no):
        if no in self.conex:
            return 
        if no == self:
            return 
        self.conex.append(no)
        no.conex.append(self)

     def desconecta(self, no, lista):
         self.conex.remove(no)
         no.conex.remove(self)
         if len(no.conex) < 1:
             lista.remove(no)

         if len(self.conex) < 1:
             lista.remove(self)

     def deleta_no(self,lista):
         for no in self.conex:
             no.conex.remove(self)
         lista.remove(self)
        
    


nos = []
for i in range(0,10):
    nos.append(No(i*8, i*2, i))

nos[0].conecta_no(nos[1])
nos[1].conecta_no(nos[2])
nos[1].conecta_no(nos[3])
nos[3].conecta_no(nos[4])
nos[2].conecta_no(nos[4])
nos[2].conecta_no(nos[5])
nos[5].conecta_no(nos[6])
nos[5].conecta_no(nos[7])
nos[6].conecta_no(nos[7])
nos[7].conecta_no(nos[8])
nos[8].conecta_no(nos[9])

for no in nos:
    print()
    print(no.info,end="")
    for conexao in no.conex:
        print("-", conexao.info,end="")


nos[2].deleta_no(nos)
print()
for no in nos:
    print()
    print(no.info,end="")
    for conexao in no.conex:
        print("-", conexao.info,end="")
```
