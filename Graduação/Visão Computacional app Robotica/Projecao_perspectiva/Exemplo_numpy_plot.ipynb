## Manipulação de Vetores

Os comandos a seguir podem ser usados para inicializar vetores utilizando a biblioteca `numpy`.

Para inicializar um vetor com elementos pré-definidos:

```python
a = np.array([1, 3, 7, 8])
a
```

Para inicializar uma matriz com elementos pré-definidos:

```python
A = np.array([[1, 5, 3, 9],
              [0, 2, 0, 2]])
A
```

Para inicializar um vetor ou matriz com zero ou um:

```python
N = 2
M = 4
B = np.zeros((N,M)) # Matriz de 0s de dimensão NxM
C = np.ones((N,M)) # Matriz de 1s de dimensão NxM
print("B = ",B)
print("C = ",C)
```

Para inicializar um vetor de elementos igualmente espaçados por `step`, que começa em `start` e vai até `end` (excluindo o valor `end`):

```python
start = 0
step = 0.5
end = 10+step # Observe que para que o vetor termine no valor 10, é preciso definir end=10+step
t = np.arange(start,end,step)
t
```

Para inicializar um vetor de `num` elementos igualmente espaçados que começa em `start` e vai até `end` (incluindo o valor `end`):

```python
start = 0
end = 10
num = 21    # Observe que para conseguir o mesmo step que o exemplo anterior, num=(1+(end-start)/step)
t = np.linspace(start,end,num)
print("t = ",t)
print("t.shape =",t.shape)
```

As operações entre vetores ou matrizes: adição, subtração, multiplicação, divisão, exponenciação, etc, são realizadas elemento a elemento dos vetores.

```python
print("a = ",a)
b = a + a
print("b = a+a = ",b)
c = b**2 # Aqui é feita a exponenciação de b por 2, ou seja, c=b*b
print("c = b**b = ",c)
d = c/a
print("d = c/a = ",d)
```

Nas operações envolvendo escalares e vetores, ou escalares e matrizes, a operação com o escalar é feita para todos os elementos do vetor ou matriz:

```python
print("a = ",a)
f = 2*a - 1
print("f = ",f)
```

Cuidado com a prioridade de uma operação em relação a outra. Na dúvida utilize parênteses para deixar claro a ordem em que as operações devem ser feitas:

```python
x = 3*5**2 # Aqui a operação de exponenciação é realizada primeiro, depois a multiplicação
print("x = ",x)
y = (3*5)**2 # Aqui a operação de exponenciação é realizada no resultado da multiplicação
print("y = ",y)
```

Para fazer a multiplicação matricial, utilize o comando `np.dot`

No exemplo a seguir, um vetor $N \times 1$ é multiplicado por um vetor $1 \times M$ para gerar uma matriz $N \times M$:

```python
print("a = ",a)
print("Dimensão ou shape do vetor a: ",a.shape,"\n")
print("t = ",t)
print("Dimensão ou shape do vetor t: ",t.shape,"\n")
dim_a = len(a)
dim_t = len(t)

a = np.expand_dims(a, axis=1) # Transforma o vetor unidimensional a em um vetor coluna
t = np.expand_dims(t, axis=0) # Transforma o vetor unidimensional a em um vetor linha

print("Dimensão ou shape do vetor a: ",a.shape,"\n")
print("Dimensão ou shape do vetor t: ",t.shape,"\n")


At = np.dot(a,t) # Não é possível fazer a transposta de um vetor unidimensional, é preciso usar o comando reshape
print("At = ",At)
print("Dimensão ou shape do vetor At: ",At.shape,"\n")

print("A = ",A)
print("Dimensão ou shape da matriz A: ",A.shape,"\n")
print("C = ",C)
print("Dimensão ou shape da matriz C: ",C.shape,"\n")

ATC = np.dot(A.T,C) # Realiza a operação (A^T)C, ou seja, transposta de A multiplicada por C
print("ATC = ",ATC)
print("Dimensão ou shape de ATC: ",ATC.shape,"\n")

CTA = np.dot(C.T,A) # Realiza a operação (C^T)A, ou seja, transposta de C multiplicada por A
print("CTA = ",CTA)
print("Dimensão ou shape de CTA: ",CTA.shape,"\n")

CAT = np.dot(C,A.T) # Realiza a operação C(A^T), ou seja, C multiplicado pela transposta de A
print("CAT = ",CAT)
print("Dimensão ou shape de CAT: ",CAT.shape,"\n")

ACT = np.dot(A,C.T) # Realiza a operação A(C^T), ou seja, A multiplicado pela transposta de C
print("ACT = ",ACT)
print("Dimensão ou shape de ACT: ",ACT.shape,"\n")

```

Para acessar elementos específicos, colunas, linhas, ou partes das matrizes:

```python
print("A = ", A, "\n")

As1 = A[:,2] # Acessa a terceira coluna de A
print("A[:,2] =",As1, "\n")

As2 = A[1,:] # Acessa a segunda linha de A
print("A[1,:] =", As2, "\n")

As3 = A[:,1:4] # Acessa da segunda coluna até a quarta coluna de A
print("A[:,1:4] =", As3, "\n")
```

## Plotar um gráfico

O exemplo a seguir gera um gráfico de duas curvas sobrepostas com legendas

```python
t_cpy = t.flatten() #é necessario pois t é uma matriz (1,21)
y1 = 10*t_cpy**2 + 5*t_cpy + 2
y2 = 5*t_cpy**2 + 4*t_cpy

print(y1)
print(y2)
```

O exemplo anterior também poderia ser executado utilizando o seguinte código que considera vetores de coeficientes para as equações polinomiais:

$$ \bf{y} = \bf{c}_2 \bf{t}^2 + \bf{c}_1 \bf{t}^2 + \bf{c}_0 $$

```python
c2 = np.array([10,5])
print(c2.shape)
c2 = np.expand_dims(c2, axis=1)
print("Shape do vetor c2: ", c2.shape)

c1 = np.array([5,4])
c1 = np.expand_dims(c1, axis=1) #parametro axis adiciona uma dimensão ao vetor/matriz
print("Shape do vetor c1: ", c1.shape)

c0 = np.array([2,0])
c0 = np.expand_dims(c0, axis=1)
print("Shape do vetor c0: ", c0.shape)

print("Shape do vetor t: ", t.shape)

y = np.dot(c2,t**2) + np.dot(c1,t) + c0

print("Shape do vetor y: ", y.shape)

plt.plot(t.T,y.T)  # Precisa transpor para que o vetor de tempo seja vetor coluna
plt.ylabel('Amplitude (V)')
plt.xlabel('Tempo (s)')
plt.legend(['y1','y2'])
plt.grid(True)
plt.axis([0,10,0,1000])
```

## Exercício 1

Plote na mesma figura as duas curvas a seguir em função do tempo ($t$):

$$ y_1(t) = A_1 \sin(2 \pi f t) $$
$$ y_2(t) = A_2 \cos(2 \pi f t + \pi/4) $$

Considere os seguintes valores: $A_1 = 2$, $A_2 = 3$, $f = 2$. Plote para o intervalo de tempo começando e $t=0$ e terminando em $t=2$ (incluindo este último instante de tempo).

```python
#constantes
a1 = 2
a2 = 3
f = 2
start = 0
step = 0.01
end = 2+step

t = np.arange(start, end, step) #vetor tempo
#funções y1 e y2 em função do tempo
y1 = a1*np.sin(2*pi*f*t)
y2 = a2 * np.cos(2*pi*f*t+(pi/4))

#plotando os graficos
plt.plot(t, y1, 'purple', t, y2, 'r')
plt.ylabel('Amplitude (V)')
plt.xlabel('Tempo (s)')
plt.legend(['y1','y2'])
plt.grid(True)
plt.axis([0, end, -4, 4])
plt.show()

```

## Exercício 2

Faça um plot tridimensional da seguinte curva paramétrica em função do tempo, onde as coordenadas de um ponto $(x,y,z)$ são dadas pelas equações:

$$ x = 3 \cos(2\pi f t) $$
$$ y = 3 \sin(2 \pi f t) $$
$$ z = t $$

Considere que $f = 2$, e o intervalo de tempo começando e $t=0$ e terminando em $t=3$ (incluindo este último instante de tempo).

Dica: Consulte a documentação do Matplotlib: https://matplotlib.org/stable/gallery/mplot3d/lines3d.html

```python
#vetor tempo
start = 0
step = 0.01
end = 3+step
t = np.arange(start, end, step)
f = 2
#variaveis x,y,z
theta = 2*pi * f* t
x = 3 * np.cos(theta)
y = 3 * np.sin(theta)
z = t

#plotando a figura
ax = plt.figure().add_subplot(projection='3d')

ax.plot(x,y,z,color="purple", label="Curva Paramétrica" )
ax.legend()
ax.grid(True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
