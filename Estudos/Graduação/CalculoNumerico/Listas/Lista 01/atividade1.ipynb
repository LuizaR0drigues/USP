```python
import numpy as  np
import matplotlib.pyplot as plt
import time


                      
def CalcArray(n, alpha, beta):
    a = np.random.rand(n)
    b = np.random.rand()
    ti = time.time()
    c = alpha * a + beta * b
    tf = time.time() - ti
    return c, tf
def TempoPlot(dimens, alpha, beta):
    linscale = []
    logscale = []

    for n in dimens:
        c, tempo = CalcArray(n, alpha, beta)
        linscale.append(tempo)
        logscale.append(np.log(tempo))
        
    fig, ax = plt.subplots(1,2,figsize=(20,10))

    ax[0].plot(dimens, linscale, '-r')
    ax[0].grid()
    ax[0].set_xlabel('Dimensão(n)')
    ax[0].set_ylabel('Tempo')

    ax[1].loglog(dimens, linscale)
    ax[1].grid()
    ax[1].set_xlabel('Dimensão(n)')
    ax[1].set_ylabel('Tempo(log)')

    plt.show()


dimens= [10**2,10**3,10**4,10**5,10**6,10**7,10**8]
alpha = np.random.rand()
beta = np.random.rand()

TempoPlot(dimens, alpha, beta)

```

```python
import numpy as np
import matplotlib.pyplot as plt

def calc_a(a):
    N = 5000
    x = 0.1
    vet = []

    for i in range(1,N):
        vet.append(x)
        x = a* x *(1-x)
    return vet
    
def media(vetor):
    N= 5000
    
    media = soma(vetor)/N
    return media
    
def variancia(vetor):
    N= 5000
    variancia =0
    variancia = (1/(N-1))*soma((np.square([x - media(vetor) for x in vetor])))
    #print(f'var: {variancia}')
    #print("fomart ", np.var(vetor,ddof=1))
    return variancia

def soma(vetor):
    
    soma =0
    for x in vetor:
        soma += x
    return soma


def plot(a):
    x = a
    Y =[]
    v2 = []
    v3 = []
    v4 = []
    for z in a:
        vet = calc_a(z)
        Y.append(media(vet))
        v2.append(variancia(vet))
        v3.append(np.var(vet))
        v4.append(np.median(vet))
        
    fig, ax = plt.subplots(2,2, figsize=(10,6))
    
    ax[0,0].plot(x,Y,"-r")
    ax[0,0].legend(["Função manual"])
    ax[0,0].grid()
    ax[0,0].set_ylabel("média")
    
    ax[0,1].plot(x,v2,"-g")
    ax[0,1].legend(["Função manual"])
    ax[0,1].grid()
    ax[0,1].set_ylabel("variança")

    ax[1,0].plot(x,v4,"-r")
    ax[1,0].legend(["Função Numpy"])
    ax[1,0].grid()
    ax[1,0].set_ylabel("média")
    
    ax[1,1].plot(x,v3,"-g")
    ax[1,1].legend(["Função Numpy"])
    ax[1,1].grid()
    ax[1,1].set_ylabel("Variança")
    plt.show()


#a =[1,2,3.8,4]
a = np.linspace(0,4,100)
plot(a) 
#print(media(calc_a(a)))
#print(variancia(calc_a(a)))
    
```
